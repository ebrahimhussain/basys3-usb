```python
DRIVE = 12
SLEW = "FAST"

pins = []

while True:
    name = input("pin name: ")
    if name == "done":
        break

    pin = input("pin: ")
    pins.append((name, pin))

for name, pin in pins:
    print(f"set_property DRIVE {DRIVE} [get_ports {name}]")

for name, pin in pins:
    print(f"set_property SLEW {SLEW} [get_ports {name}]")

for name, pin in pins:
    print(f"set_property PACKAGE_PIN {pin} [get_ports {name}]")
```
